```python
%matplotlib inline
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_array = misc.imread('potato.jpg')
```

```python
plt.imshow(img_array)
```

```python
def average(pixel):
    return (pixel[0] + pixel[1] + pixel[2])/3
```

```python
grey = np.zeros((img_array.shape[0], img_array.shape[1]))
for rownum in range(len(img_array)):
   for colnum in range(len(img_array[rownum])):
      grey[rownum][colnum] = average(img_array[rownum][colnum])
```

```python
plt.imshow(grey)
```

```python
print(img_array.shape)
print(grey.shape)
```

```python
def weighted_average(pixel):
    return 0.299*pixel[0] + 0.587*pixel[1] + 0.114*pixel[2]
```

```python
grey_wt = np.zeros((img_array.shape[0], img_array.shape[1]))
for rownum in range(len(img_array)):
   for colnum in range(len(img_array[rownum])):
      grey_wt[rownum][colnum] = weighted_average(img_array[rownum][colnum])
```

```python
plt.imshow(grey_wt)
```
