```python
%matplotlib inline

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("peacock.jpg")
mask = np.zeros(img.shape, dtype = "uint8")
x1, y1 = 250, 150
x2, y2 = 350, 350
cv2.rectangle(mask, (x1, y1), (x2, y2), (255, 255, 255), -1)
maskedImg = cv2.bitwise_and(img, mask)
plt.imshow(maskedImg)
```

```python
l = np.absolute(y1-y2)
b = np.absolute(x1-x2)
```

```python
print("Length : ", l)
print("Width : ", b)
```
